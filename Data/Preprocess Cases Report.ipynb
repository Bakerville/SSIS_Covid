```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Cases Report.csv")
```

```python
data
```

```python
def preprocess(row):
    li = row["Outcome"].split(",")
    index = 0
    if(len(li)>1):
        for name in row.index:
            row[name] = li[index]
            index = index + 1
    return row
```

```python
data=data.apply(preprocess, axis="columns")
```

```python
data
```

```python
data.loc[data["CaseReported Date"] ==  ' Pine Ridge District Health Unit"',:]
```

```python
def specified_process(row):
    row["Reporting PHU"] = row["Reporting PHU"] +  row["SpecimenDate"] + row["CaseReported Date"]
    row["Reporting PHU"] = row["Reporting PHU"][1:-1]
    col = data.columns
    for idx in range(4,len(col)-2):
        row[row.index[idx]] =  row[row.index[idx+2]]
    row[-1] = np.nan
    row[-2] = np.nan
    return row
```

```python
data.loc[data["CaseReported Date"] ==  ' Pine Ridge District Health Unit"',:]=data.loc[data["CaseReported Date"] ==  ' Pine Ridge District Health Unit"',:].apply(specified_process, axis="columns")
```

```python
data
```

```python
def specified_process_2(row):
    row["Reporting PHU"] = row["Reporting PHU"] +  row["SpecimenDate"]
    row["Reporting PHU"] = row["Reporting PHU"][1:-1]
    col = data.columns
    for idx in range(4,len(col)-1):
        row[row.index[idx]] =  row[row.index[idx+1]]
    row[-1] = np.nan
    return row

data.loc[(data["SpecimenDate"] ==  ' Frontenac and Lennox & Addington Public Health"') | (data["SpecimenDate"] ==  ' Public Health"') | (data["SpecimenDate"] ==  ' Grenville and Lanark District Health Unit"') ,:]=data.loc[(data["SpecimenDate"] ==  ' Frontenac and Lennox & Addington Public Health"') | (data["SpecimenDate"] ==  ' Public Health"') | (data["SpecimenDate"] ==  ' Grenville and Lanark District Health Unit"') ,:].apply(specified_process_2, axis="columns")
```
